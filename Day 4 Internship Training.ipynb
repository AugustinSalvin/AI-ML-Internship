```python
df['Department_mode_imputed'] = df['Department'].fillna(df['Department'].mode()[0])#counts from first index onwards and replace NuN  value by most repeated value/term
df['Department_mode_imputed']
```

```python
numerical_features=df.select_dtypes(include=['float64','int64'])
skewness=numerical_features.skew().sort_values(ascending=False)
print("Skewness of Numerical Features:")
print(skewness)
```

```python
# Plot distributions of numerical features
import matplotlib.pyplot as plt
plt.figure(figsize=(25, 20))
for i, column in enumerate(numerical_features.columns, 1):
    plt.subplot((len(numerical_features.columns) // 3) + 1, 3, i)
    sns.histplot(df[column].dropna(), kde=True, bins=30, color='blue')
    plt.title(f'{column} (Skewness: {skewness[column]:.2f})', fontsize=10)
    plt.xlabel(column, fontsize=8)
    plt.ylabel('Frequency', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
company_data=pd.read_csv("/content/preprocessing_practice_dataset.csv")

# Select numerical columns for scaling
numerical_columns = ['Age', 'Monthly_Salary', 'Department']

# Drop rows with missing values for simplicity
company_data.dropna(subset=numerical_columns, inplace=True)

# Initialize scalers
minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

# Apply Min-Max Scaling
company_data['Age_MinMax'] = minmax_scaler.fit_transform(company_data[['Age']])
company_data['Salary_MinMax'] = minmax_scaler.fit_transform(company_data[['Monthly_Salary']])

# Apply Standard Scaling
company_data['Age_Standard'] = standard_scaler.fit_transform(company_data[['Age']])
company_data['Salary_Standard'] = standard_scaler.fit_transform(company_data[['Monthly_Salary']])


# Display scaled features
company_data[['Age', 'Age_MinMax', 'Age_Standard','Salary_MinMax','Salary_Standard']].head()
```
