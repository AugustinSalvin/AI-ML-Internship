Pclass: This refers to the passenger class (1st, 2nd, or 3rd class). It's a proxy for socio-economic status, with 1st class being upper, 2nd middle, and 3rd lower class.
SibSp: This indicates the number of siblings or spouses aboard the Titanic with the passenger. (Sibling = brother, sister, stepbrother, stepsister; Spouse = husband, wife (mistresses and fiancés were ignored)).
Parch: This indicates the number of parents or children aboard the Titanic with the passenger. (Parent = mother, father; Child = daughter, son, stepdaughter, stepson. Some children travelled only with a nanny, therefore Parch=0 for them.)
Embarked: This denotes the port of embarkation – where the passenger got on the ship. The typical codes are:
'C' = Cherbourg
'Q' = Queenstown
'S' = Southampton

# **Missing value imputation**

```python
fd['Age_mean_imputed'] = fd['Age'].fillna(fd['Age'].mean())
fd['Age_mean_imputed']
```

```python
fd['Age_median_imputed'] = fd['Age'].fillna(fd['Age'].median())
fd['Age_median_imputed']
```

```python
fd['Cabin_mode_imputed'] = fd['Cabin'].fillna(fd['Cabin'].mode()[0])#counts from first index onwards and replace NuN  value by most repeated value/term
fd['Cabin_mode_imputed']
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
#scikitlearn
# Load the dataset
file_path = '/content/titanic uncleaned (1).csv'
titanic_data = pd.read_csv(file_path)

# Select numerical columns for scaling
numerical_columns = ['Age', 'Fare', 'Pclass']

# Drop rows with missing values for simplicity
titanic_data.dropna(subset=numerical_columns, inplace=True)

# Initialize scalers
minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

# Apply Min-Max Scaling
titanic_data['Age_MinMax'] = minmax_scaler.fit_transform(titanic_data[['Age']])
titanic_data['Fare_MinMax'] = minmax_scaler.fit_transform(titanic_data[['Fare']])

# Apply Standard Scaling
titanic_data['Age_Standard'] = standard_scaler.fit_transform(titanic_data[['Age']])
titanic_data['Fare_Standard'] = standard_scaler.fit_transform(titanic_data[['Fare']])


# Display scaled features
titanic_data[['Age', 'Age_MinMax', 'Age_Standard']].head()
```
