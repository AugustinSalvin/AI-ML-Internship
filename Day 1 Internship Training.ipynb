Write a function that accepts a string and returns the number of vowels in it.

```python
s = input()
def vowels(s):
    count = 0
    for ch in s:
        if ch in "aeiouAEIOU":
            count += 1
    return count
print(vowels(s))
```

Write a function that accepts a number and returns True if it is prime, otherwise False.

```python
x = int(input())
def prime(x):
    if x < 2:
        return False
    for i in range(2, x):
        if x % i == 0:
            return False
    return True
print(prime(x))
```

Write a function that accepts a string and a character and returns how many times that character appears.

```python
x = input()
s = input()
def chara(s):
  count = 0
  for ch in s:
    if ch == x:
      count += 1
  return count
print(chara(s))
```

```python
x = input()
s = input()
def chara(s):
  print(s.count(x))
chara(s)
```

Write a function that returns how many odd digits are present in a number.

```python
x = input()
def odd(x):
  count = 0
  for i in x:
    if int(i) % 2 == 1:
      count += 1
  return count
print(odd(x))
```

Write a function that accepts a string and returns the string without any spaces.

```python
s = input()
def no_space(s):
  return s.replace(" ","")
print(no_space(s))
```

Write a function that accepts three numbers and returns the smallest one.

```python
x = input()
def Small(x):
  small = x[0]
  for i in x:
    if int(i) < int(small):
      small = i
  return small
print(Small(x))
```
